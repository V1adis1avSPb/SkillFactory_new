```python
import pandas as pd
import numpy as np
from sklearn import metrics
from sklearn import linear_model
from sklearn import tree
from sklearn import ensemble
```

```python
df = pd.read_csv('data/weatherAUS.csv')
df.info()
```

```python
df.head(20)
```

```python
df.isnull().sum().sum()
```

```python
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
df = df.dropna(axis=1, thresh=0.6*len(df))
df.info()
```

```python
df['RainToday'] = df['RainToday'].apply(lambda x: 1 if x == 'Yes' else (0 if x == 'No' else x))
df['RainTomorrow'] = df['RainTomorrow'].apply(lambda x: 1 if x == 'Yes' else (0 if x == 'No' else x))
```

```python
round(df['RainToday'].mean(),2)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.info()
```

```python
df['Month'] = df['Date'].apply(lambda x: x.month)
df = df.drop('Date', axis=1)
```

```python
df.groupby(['Month'])['RainToday'].agg('mean')
```

```python
df.info()
```

```python
categoricals = ['Month', 'Location', 'WindGustDir', 'WindDir9am', 'WindDir3pm']
```

```python
df = pd.get_dummies(df, columns=categoricals)
```

```python
df = df.dropna(axis=0)
```

```python
df.info()
```

```python
from sklearn import preprocessing
from sklearn import model_selection
X = df.drop('RainTomorrow', axis=1)
y = df['RainTomorrow']
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.3, random_state=31)
```

```python
round(y_test.mean(),2)
```

```python
X_train['MinTemp']
```

```python
np.random.seed(31)
mean_lst = []
for i in range(0,1000):
    a = np.random.randint(X_train['MinTemp'], len(X_train))
    mean_lst.append(X_train.iloc[a]['MinTemp'].mean())
mean_lst
```

```python
import statistics
round(statistics.stdev(mean_lst),2)
```

```python
log = linear_model.LogisticRegression()
log.fit(X_train, y_train)
y_pred = log.predict(X_test)
round(metrics.roc_auc_score(y_test, y_pred),2)
```

```python
tr_cl = tree.DecisionTreeClassifier(random_state=42)
params = {'max_leaf_nodes': list(range(2, 10)), 'min_samples_split': [2, 3, 4], 'max_depth': [5,7,9,11]}
grid = model_selection.GridSearchCV(estimator=tr_cl, param_grid=params, cv=3)
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
y_pred = grid.predict(X_test)
metrics.roc_auc_score(y_test, y_pred)
```

```python
rf = ensemble.RandomForestClassifier(max_depth=15, max_features=7, min_samples_leaf=3)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
print(round(metrics.roc_auc_score(y_test, y_pred),2))
s = pd.DataFrame(rf.feature_importances_, X_train.columns, columns=['importance'])
s.sort_values('importance', ascending=False)
```

```python
rf = ensemble.RandomForestClassifier(n_estimators=100, random_state=31)
params = {'max_features': [ 4, 5, 6, 7], 'min_samples_leaf': [3, 5, 7, 9, 11], 'max_depth': [5, 10, 15]}
grid = model_selection.GridSearchCV(estimator=rf, param_grid=params, cv=3)
grid.fit(X_train, y_train)
y_pred = grid.predict(X_test)
round(metrics.roc_auc_score(y_test, y_pred),2)
```
